```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#checking for null value
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Shaadi'}) 
```

```python
# describe() function used for description of data in dataframe (i.e count, mean, std, etc)
df.describe()
```

```python
#used for description on specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
#Exploratory Data Analysis
```

```python
# plotting a bar chart for gender and it's count
palette = ['#3498db', '#e74c3c'] 
# Create the countplot with 'hue' set to 'Gender'
ax = sns.countplot(x='Gender', data=df, hue='Gender', palette=palette, legend=False)
for bars in ax.containers:
    ax.bar_label(bars, color='black') 
plt.show()



```

```python
#plotting bar graph for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, )
plt.show()
```

```python
# plotting graph on the bases of age group with respect to gender
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
custom_colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta']
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, palette=custom_colors, hue='Age Group', legend=False)

plt.show()

```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
#Total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
custom_colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#FF8C33', '#33FFF7', '#8E44AD', '#F39C12', '#2ECC71', '#D35400']
sns.barplot(data=sales_state, x='State', y='Orders', hue='State', palette=custom_colors, legend=False)
plt.show()

```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

```python
sns.set(rc={'figure.figsize':(20,5)})
custom_colors = [
    'red', 'green', 'blue', 'purple', 'orange', 
    'cyan', 'magenta', 'yellow', 'brown', 'pink', 
    'gray', 'lightblue', 'lime', 'teal', 'indigo'
]
ax = sns.barplot(data=sales_state, x='Occupation', y='Amount', 
                 hue='Occupation', palette=custom_colors, legend=False)
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
sns.set(rc={'figure.figsize':(30,3)})
custom_colors = [
    'red', 'green', 'blue', 'purple', 'orange', 
    'cyan', 'magenta', 'yellow', 'brown', 'pink', 
    'gray', 'lightblue', 'lime', 'teal', 'indigo',
    'darkred', 'darkgreen', 'darkblue'
]
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', 
                   palette=custom_colors, legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Top 10 Products Sold Categorywise

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
custom_colors = [
    'red', 'green', 'blue', 'purple', 'orange', 
    'cyan', 'magenta', 'yellow', 'brown', 'pink'
]
ax = sns.barplot(data=sales_state, x='Product_Category', y='Amount', palette=custom_colors, hue='Product_Category', legend=False)

  

```

*From above graphs we can see that most sold products are from Food, Clothing and Electronics categorywise*
